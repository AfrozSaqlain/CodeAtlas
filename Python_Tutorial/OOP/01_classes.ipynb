# Why do we need classes and objects?

Suppose I want to maintain a register for different classes in a school. I want to use a single template whose functionality would be:
- To store names of students in a class.
- To store their ages.
- To store how much total marks they have gotten.
- And I want to be able to compute different metrics, like class average, top 3 rankers, etc.

To do all of this either we will have to define a variable for each and every student from every class, but that will become too chaotic so instead we use classes and objects.

`class` is a user-defined datatype that allows us to create templates for organizing data and functionality. An instance of a class is called an `object`. In our school example, the template that defines how student information is stored and processed is the `class`. By itself, the `class` only describes the structure and behavior. It becomes useful when we create objects from it, such as separate registers for Class 10A, Class 10B, and so on, each maintaining its own set of students and records while following the same template.

We define a name for the class which can be anything.

Next is we define an `__init__` method, which you must add as the very first thing after defining a class. This is the function which automatically runs when you create an instance of a class (which we did in the 2nd next cell). 

```python
class ClassOfStudents:
    def __init__(self):
        self.name = []
        self.age = []
        self.marks = []

    def add_biodata(self, biodata):
        self.name.append(biodata[0])
        self.age.append(biodata[1])
        self.marks.append(biodata[2])
```

```python
class1 = ClassOfStudents()
```

```python
class1.add_biodata(biodata=['Raj', 23, 524])
```
